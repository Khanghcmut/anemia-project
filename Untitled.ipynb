```python
from sklearn.linear_model import LogisticRegression #logistic regression
from sklearn import svm #support vector Machine
from sklearn.ensemble import RandomForestClassifier #Random Forest
from sklearn.neighbors import KNeighborsClassifier #KNN
from sklearn.naive_bayes import GaussianNB #Naive bayes
from sklearn.tree import DecisionTreeClassifier #Decision Tree
from sklearn.model_selection import train_test_split #training and testing data split
from sklearn import metrics #accuracy measure
from sklearn.metrics import confusion_matrix #for confusion matrix
from xgboost import XGBClassifier#XGBcalssifier
```

```python
x=df.iloc[:,:-1]
y=df.Result
x.shape
```

```python
xtrain,xtest, ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
model=svm.SVC(kernel='rbf',C=1,gamma=0.1)
model.fit(xtrain,ytrain)
prediction1=model.predict(xtest)
print('Accuracy for rbf SVM is ',metrics.accuracy_score(prediction1,ytest))
```

```python
decision_tree = DecisionTreeClassifier (criterion = 'entropy', max_depth = 10, random_state=2000)
decision_tree.fit(xtrain,ytrain)
decision_tree.score(xtest, ytest)
```

```python
from sklearn.model_selection import StratifiedKFold #for K-fold cross validation
from sklearn.model_selection import cross_val_score #score evaluation
from sklearn.model_selection import cross_val_predict #prediction
seed=200
kfold = StratifiedKFold(10, shuffle=True, random_state=seed) # k=10, split the data into 10 equal parts

cv_mean=[]
accuracy=[]
std=[]
classifier=[]
models=[svm.SVC(kernel='linear'),LogisticRegression(),KNeighborsClassifier(n_neighbors=9),DecisionTreeClassifier(random_state=seed),GaussianNB(),RandomForestClassifier(random_state=seed),XGBClassifier (use_label_encoder=False, eval_metric='logloss', random_state=seed)]
for model in models:
    model_name = model.__class__.__name__
    if model_name not in(classifier):
        classifier.append(model_name)
    cv_result = cross_val_score(model,x,y, cv = kfold,scoring = 'average_precision')
    cv_result=cv_result
    cv_mean.append(round(cv_result.mean(),2))
    std.append(round(cv_result.std(),2))
    accuracy.append(cv_result)
new_models_dataframe=pd.DataFrame({'ML model':classifier,'accuracy_score':cv_mean,'std':std})
new_models_dataframe
```

```python
entries=[]
for model in models:
    model_name = model.__class__.__name__
    kfold = StratifiedKFold(50, shuffle=True, random_state=seed)
    scores = cross_val_score(model, x,y, scoring='average_precision', cv=kfold)
    for fold_idx, score in enumerate(scores):
        entries.append([model_name, fold_idx, score])
cv_df = pd.DataFrame(entries, columns= [ 'model_name','fold id','accuracy_score' ])
sns.boxplot(cv_df,x='model_name',y='accuracy_score',showmeans=True)
plt.xticks(rotation=60)
cv_df
```
